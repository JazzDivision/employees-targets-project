```python
# Import pandas library for data manipulation
import pandas as pd
```

```python
# Read the CSV file
employees = pd.read_csv("employees.csv", dayfirst=True)
employees.head()
```

```python
# Print whole dataset
employees
```

```python
# Check for duplicates
employees[employees.duplicated("employee_id")]

# No duplciate employee_id entries were identifed
```

```python
# Count rows and columns to check for expected number
employees.shape
```

```python
# Check the schema for datatypes inconsistencies
employees.info()
```

```python
# Parse start_date as a real date, not object
employees["start_date"] = pd.to_datetime(
    employees["start_date"],
    dayfirst=True, # (dd/mm/yyyy) format
    errors="coerce" # Valid dates are parsed, bad dates = NaT
)
```

```python
# Re-check the data types to ensure data has been correctly parsed
employees.info()
```

```python
# Save the clean CSV
employees.to_csv("clean_employees.csv", index=False)
```
