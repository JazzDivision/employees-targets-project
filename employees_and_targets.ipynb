```python
# Import pandas library for data manipulation
import pandas as pd
```

```python
# Read the CSV file
employees = pd.read_csv("employees.csv", dayfirst=True)
employees.head()
```

```python
# Check for duplicates
employees[employees.duplicated("employee_id")]
```

```python
# Check content of bottom 10 rows to confirm empty rows
employees.tail(10)
```

```python
# Check to see how many rows are empty
employees.isna().all(axis=1).sum()
```

```python
# Remove rows where every column is null
employees = employees.dropna(how="all")
```

```python
# Check that duplicates were dropped
employees.shape
```

```python
# Check the schema for datatypes
employees.info()
```

```python
# Convert employee_id to an integer
employees["employee_id"] = employees["employee_id"].astype(int)
employees.info()
```

```python
# Parse start_date as a real date, not object
employees["start_date"] = pd.to_datetime(
    employees["start_date"],
    dayfirst=True, # (dd/mm/yyyy) format
    errors="coerce" # Valid dates are parsed, bad dates = NaT
)

employees.info()
```

```python
# Save the clean CSV
employees.to_csv("clean_employees.csv", index=False)
```
